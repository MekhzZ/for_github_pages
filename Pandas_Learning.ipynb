# Filtering data



1.   using labels
2.   using position
3.   using boolean conditions



# using labels to filter

```python
#filtering columns using labels
#single column

df.Gender
```

```python
#multiple columns
df[['Gender','Height']]
```

```python
#filtering rows using labels

#single row
df.loc[0] #first row data display
```

```python
#multiple rows
df.loc[[0,1,2]]
```

```python
#filtering rows and columns together

df.loc[0:5,['Gender','Height','Weight']]
```

# using positions to filter

```python
#filtering rows using position

#single row

df.iloc[0]
```

```python
#multiple rows
df.iloc[[0,1,2]]
```

```python
df.head()
```

```python
df.iloc[:,[2]] # all rows data of column 2
```

```python
#filtering rows and columns together using position

df.iloc[[0,1],[1,2]]
```

# using boolean conditions to filter

```python
df.loc[df.Height>170]
```

```python
df.value_counts('Gender')
```

```python
df.loc[(df.Gender == 'Male') & (df.Height > 170)]
```
